### 01) WAP to find sum of all the elements in a List.

```python
s=int(input("enter elements:"))
l1=[]
for i in range(0,s):
    l1.append(int(input("enter list element:")))
sum=0
for i in l1:
    sum =sum + i
print(sum)    
```

### 14) WAP to extract elements with frequency greater than K.

```python
l=list(map(int,input("enter list:").split()))
k=int(input("enter frequency:"))
r=[]
s=[]
for i in l:
    if l.count(i) > k:
        r.append(i)
        if r.count(i)==1:
            s.append(i)
print(s)
```

### 17) WAP to create a list of common elements from given two lists.

```python
l1=list(map(int,input("enter list1:").split()))
l2=list(map(int,input("enter lis2t:").split()))
l3=[]
for i in l1:
    for j in l2:
        if i==j:
            l3.append(j)
print(l3)            
```
