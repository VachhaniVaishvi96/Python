### 14) WAP to check whether the given number is Palindrome or not.

```python
n=int(input("enter number:"))
og=n
ans=0
while og != 0:
    rem = og%10
    ans= ans*10+rem
    og = og//10
if(n == ans):
    print("is Palindrom")
else:
    print("not")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
n=int(input("enter number:"))
og=n
count=0
sum=0
while og != 0:
    count=count+1
    og = og//10
og=n
while og != 0:
    rem=og%10
    pow=1
    for i in range(1,count+1):
        pow = pow*rem
    sum=sum+pow
    og=og//10

if sum == n:
    print("armstrong")
else:
    print("not")
```
