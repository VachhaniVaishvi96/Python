 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
a='Python3@Lab'
b='HELLO123'
c='hi@#9'
upper=0
lower=0
digit=0
special=0
for i in c:
    if i.islower():
        lower=lower+1
    elif i.isupper():
        upper=upper+1
    elif i.isdigit():
        digit+=1
    else:
        special=special+1
print(upper)
print(lower)
print(digit)
print(special)
```

### 9) String Compression
- `aaabbcccc` → `a3b2c4`
- `xxxyyyz` → `x3y3z1`
- `abcd` → `a1b1c1d1`

```python
s=input("enter string:")
ans=''
c=1
for i in range(1,len(s)):
    if s[i]==s[i-1]:
        c=c+1;
    else:
        ans+=s[i-1]+str(c)
        c=1
ans+=s[i-1]+str(c)
print(ans)
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
s=input("enter string")
res=''
for ch in s:
    if ch.isupper():
        res+=ch.lower()
    elif ch.islower():
        res+=ch.upper()
    else:
        res+=ch
print(res)
```

### 15) Custom Split
- `apple,banana,grape`
- `10|20|30`
- `A-B-C-D`

```python
s=input("enter string:")
d=input("delimeter:")
res=[]
word=''
for ch in s:
    if ch!=d:
        word+=ch
    else:
        res.append(word.strip())
        word=''
res.append(word.strip())
print(res)
```
