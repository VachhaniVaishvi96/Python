### 03) WAP to find tuples which have all elements divisible by K from a list of tuples.
**Sample Input:**
```
tuple_list = [(2, 4, 6), (3, 6, 9), (4, 8, 12), (5, 10, 15)]
K = 2

```
**Sample Output:**
```
Tuples divisible by 2 : [(2, 4, 6), (4, 8, 12)]
```

````python
length=int(input("enter length:"))
k= int(input("enter value of k:"))
tuple_list=[]
for i in range (0,length):
    t1=tuple(int(i) for i in input("enter element:").split(' '))
    tuple_list.append(t1)
print(tuple_list) 

ans = []
for t in tuple_list:
    flag = True
    for i in t:
        if i % k != 0:
            flag = False
            break
    if flag:
        ans.append(t)

print("Tuples divisible by", k, ":", ans)
````

### 14) WAP to create (index, value) pairs for prime numbers.

**Sample Input:**
```
T = (4, 7, 9, 11, 15)
```
**Sample Output:**
```
((1, 7), (3, 11))
```

````python
t = tuple(map(int,input("enter elements:").split()))
prime_pairs=[]
for i in range(len(t)):
    n = t[i]
    if n>1:
        is_prime = True
        for j in range(2,int(n**0.5)+1):
            if n%j == 0:
                is_prime = False
                break
        if is_prime:
            prime_pairs.append((i,n))
print(prime_pairs)
````

### 16) WAP to compute column-wise sum from a tuple of lists.

**Sample Input:**
```
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
```

**Sample Output:**
```
12 15 18

```

````python
len=int(input("enter length:"))
t=[]
for i in range (0,len):
    l1=list(int(i) for i in input("enter element:").split(' '))
    t.append(l1)
# print(t) 
t = tuple(t)
result = tuple(map(sum,zip(*t)))
print(result)
````
